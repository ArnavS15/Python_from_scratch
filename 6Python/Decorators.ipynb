```python
# Python decorators modify or enhance functions ithout changing their source code
# allows to wrap another function whose behavior needes to be extended
 # the outer function is decorator that accepts a function as an argument
 # the inner function is the the wrapper defined inside the outer function i.e the inner func calls the orignal func
# the outer function returns the inner function

def simple_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func() # last me dekho the say_hello is the func called upar func ki jaga in the brackets of simple_decorator
        print("Something is happening after the function is  called.")
    return wrapper

def say_hello():
    print("Hello!")

decorated_say_hello = simple_decorator(say_hello) # means say_hello func is executed inside simple_decorator func
decorated_say_hello()
```

```python
# @ is used to call decorators in shortcuts

def simple_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func() 
        print("Something is happening after the function is  called.")
    return wrapper

@simple_decorator
def say_hello_decorated():
    print("Hello again!")

# call the decorated function directly
say_hello_decorated()
```
