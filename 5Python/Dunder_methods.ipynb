```python
#1 : __init__ to pata hai hume
#2 : __eq__ (self,other) : Defines behavior for ==
#3 : __lt__ (self,other) : Defines behavior for < less than
#2 : __gt__ (self,other) : Defines behavior for > greater than
```

```python
class Number:
    def __init__(self, value):
        self.value = value
    
    # Define equality for number objects
    def __eq__(self, other):
        return self.value == other.value
    
    # Define less than for number objects
    def __lt__(self, other):
        return self.value < other.value

n1 = Number(10)
n2 = Number(43)
n3 = Number(11)

print(n1 == n3)
print(n1 < n2)
```

```python
# Augmented Assignment
# __iadd__(self,other): defines behavior for +=
# __isub__(self,other): defines behavior for -=

class counter:
    def __init__(self, count):
        self.count = count

    # in place addition defined:
    def __iadd__(self, other):
        self.count += other
        return self # returns self for chaining
    
    # in place subtraction defined 
    def __isub__(self, other):
        self.count -= other
        return self # returns self for chaining
    
    def __str__(self):
        return str(self.count)
c = counter(10)
c -= 5
c += 5
print(c)
```

```python
# Container/collection behavior 
class Mylist:
    def __init__(self, data):
        self.data = list(data)
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, key): # to get an item
        return self.data[key]
    
    def __setitem__(self, key, value): # to update an item value
        self.data[key] = value

    def __delitem__(self, key):
        del self.data[key]
    
    def __contains__(self, item):
        return item in self.data
    
    def __str__(self):
        return str(self.data)
    
l= Mylist([1, 2, 3, 4, 5])
print(f"length: {len(l)}")

print(f"item at index 2 : {l[2]}")
    
l[1] = 99 # updated or set value
print(f"After deleting index 1 to 99: {l}")

del l[0]
print(f"After deleting index 0: {l}")

print(f"Does 3 exist? {3 in l}")
print(f"Does 10 exist? {10 in l}")


     
```

```python
# Iteration and looping 
# making class work in loops
# __iter__(self): returns iterator
# __next__(self): gets next item like in generators

class MyRange:
    def __init__(self, start, end):
        self.current = start
        self.end = end
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current < self.end:
            num = self.current
            self.current += 1
            return num
        raise StopIteration
    
print("Iteerating Through MyRange:")
for i in MyRange(0, 5):
    print(i)
```
