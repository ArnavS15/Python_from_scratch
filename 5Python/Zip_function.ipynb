```python
# Zip function helps to unpack multiple things (like the name) if we want to zip values of 2 lists into 1 list
# example:
# list3 = will have elements as (element of lst1, element of lst2)
lst1 = ['Arnav', 'Animesh', 'Khare']
lst2 = [132,172,142]
L3 = (list(zip(lst1,lst2)))
print('Thus the new Zipped_list is:' ,  L3)
```

```python
matrix = [[1,2,3], [4,5,6], [7,8,9]]
print([list(row) for row in zip(matrix)])
print('-' * 50)
# now to transpose the matrix values use *matrix
print([list(row) for row in zip(*matrix)])
print('-' *50)
# ** matrix kareneg agar to it gives back orignal value
print([list(row) for row in zip(*[list(row) for row in zip(*matrix)])])
print('-' * 50)
```
