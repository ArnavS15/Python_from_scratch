```python
# we have a list of values i.e multiple values that we want to comprehensively define in 1 function

n = [1,2,3,2,4,3,4,3] # list
square = list(filter(lambda x : x%2==0 , n)) # list ke andar we filter jiske andar lamda is defined to its needed value,the source list (here the condition is to find multiples of 2 i.e give remainder 0 when divided by 2)
print(n)
print(square) # thus we printed only multiples of 2 in the new list
# otherwise big for loops would have been needed
```
